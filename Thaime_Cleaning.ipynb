```
def clean_text(text):

    # Remove White Space and Spacial Character
    text = re.sub(r''' #|'|"|”|“|!|-|\+|\-|\*|\%|$|\?|/|_|=|\.|>|<|\(|\)|฿|,|NaN''',' ',text)      
   
    return text
```

```
project_thaime['project_name'] = project_thaime['project_name'].apply(clean_text)
project_thaime[:5]
```

```
def tokenize_and_remove_stop_word(sentence): 
    stop_word_list = pythainlp.corpus.common.thai_stopwords() 
    word_list =  pythainlp.corpus.thai_words()
    #token_list = custom_tokenizer.word_tokenize(sentence) 
    token_list = word_tokenize(sentence , engine="newmm")
    #word_tokenize(sentence, engine="newmm")
   # token_list_wo_stop_words = [token for token in token_list if (token not in stop_word_list) 
   #                                                                 and (token not in stopword)
   #                                                                 and (token in word_list)
   #                                                                 and len(token) > 1]
    token_list_wo_stop_words = [token for token in token_list if (token not in stopword)
                                                                    and (token in word_list)
                                                                    and len(token) > 1]
    return token_list_wo_stop_words
```

```
project_thaime['TOKEN'] = project_thaime['project_name'].apply(tokenize_and_remove_stop_word)
project_thaime[:5]
```

###Extract อำเภอ ตำบล

```

# prepare address mark
level_mark = ['\s(หมู่ที่|หมู่|ม\.)', '\s(อาคาร|ตึก|หมู่บ้าน)','\s(บ้าน|หมู่บ้าน)', 
              '\s(ตำบล|ต\.|แขวง)','\s(อำเภอ|อ\.|เขต)','\s(จังหวัด|จ\.|กรุงเทพมหานคร|กทม\.|กรุงเทพฯ)']
level_all = '|'.join(level_mark)
eng_level = ['Moo','Building','Alley','Tambon','Amphoe','Province']

# thai to english keyword dictionary
key_dict = dict()
for th_level, en_level in zip(level_mark, eng_level): 
    th_level_lists= th_level.split('|')
    for th in th_level_lists:
        key_dict[re.sub('[\(\)s]','',th.replace("\\",""))] = en_level

def split_frontAddrs(address):
    
    # seperate by keyword level
    addrs_compo = {}
    addrs_compo['address'] = address
    
    # find address keyword and mark '|' before the keyword
    for level in level_mark:
        if re.search(level, address):
            level_content = re.search(level, address).group()
            level_index = address.index(level_content)
            address_marked = address[:level_index] + '|' + address[level_index:]
            address = address_marked

    done_keys = [] # keep keywords that already found
    split_address = address.split('|')
    for i, split in enumerate(split_address):
        if i == 0:
            addrs_compo['frontmost_cut'] = split.strip()
        else:
            key = re.search(level_all, split).group().strip() # find keyword in each cut
            if key in ["กรุงเทพมหานคร","กทม.","กรุงเทพฯ"]: 
                keyD = key_dict['จังหวัด'] # bkk is the keyword itself, so I have to type thai key dict myself
                addrs_compo[keyD] = "กรุงเทพมหานคร" # normalize bangkok variation
            else:
                keyD = key_dict[key]
                addrs_compo[keyD] = split.replace(key, '').strip() # remove keyword, then you get address content
            done_keys.append(keyD)
    
    # find last key to be an input in the hierarchical search method
    for i, d_key in enumerate(done_keys):
        if re.search('Tambon|Amphoe|Province', d_key):
            last_key = done_keys[i - 1]
            break
        else:
            last_key = keyD    

    nan_keys = set(eng_level) - set(addrs_compo.keys()) # keep keywords that not found        
    for nan in nan_keys:
        addrs_compo[nan] = None
                
    return addrs_compo, last_key
```
